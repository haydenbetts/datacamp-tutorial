**Problem 1**: Plotting a Histogram of Iris Data

```python
# note: be mindful to name import when importing sub packages
# https://stackoverflow.com/posts/41468481/revisions
from sklearn import datasets
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# more on converting these scikit learn datasets to pandas
# https://stackoverflow.com/questions/38105539/how-to-convert-a-scikit-learn-dataset-to-a-pandas-dataset

iris_raw = datasets.load_iris()
# goal - make 150 member long array that maps 

# convert from scikit-learn bunch to pd dataframe
iris = pd.DataFrame(iris_raw.data, columns=iris_raw.feature_names)
# map iris species and their names to a column in the dataFrame
iris_species = pd.Series(iris_raw.target)
iris['species'] = iris_species.apply(lambda x: iris_raw.target_names[x])

# styling for histogram
sns.set()

plt.hist(iris['petal length (cm)'])
plt.xlabel('Count')
plt.ylabel('Petal Length (cm)')
plt.show()
```

Adjust the number of bins in histogram to be sqrt(num datapoints)

```python
import numpy as np

num_datapoints = len(iris['petal length (cm)'])
num_bins = int(np.sqrt(num_datapoints))

plt.hist(iris['petal length (cm)'], bins=num_bins)
plt.xlabel('Count')
plt.ylabel('Petal Length (cm)')
plt.show()
```

**Problem 2**: Continue exploring data using Bee swarm plot

```python
sns.swarmplot(x='species', y='petal length (cm)', data=iris)
plt.xlabel('species')
plt.ylabel('petal length (cm)')
```

**Problem 3:** Create Empirical Cumulative Distribution Function (ECDF) of Iris Data

```python
def ecdf(column):
    n = len(column)
    x = np.sort(column)
    # include extra step in second param bc arange not inclusive 
    print (1/n)
    y = np.arange(1/n, 1 + (1/n), 1/n)
    return x, y

x, y = ecdf(iris['petal length (cm)'])

plt.ylabel('ECDF')
plt.xlabel('petal length (cm)')
plt.plot(x, y, marker='.', linestyle='none')
```
